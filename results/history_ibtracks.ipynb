# IBTrACS import and visualization

With climada_env, IBTrACS histroical tracks are imported. CLIMADA already supports this. Since the subject is about India, we only need to care about North India Ocean, encoded as 'NI' in CLIMADA.

```python
from climada.hazard import TCTracks
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "cmr10"
})
plt.rcParams['font.size'] = 12
mpl.rcParams['axes.unicode_minus'] = False
```

```python
tc_tracks = TCTracks.from_ibtracs_netcdf(basin='NI')
```

```python
tracks_sync_1000yr = TCTracks.from_simulations_storm('../Data/STORM_DATA_HadGEM3-GC31-HM_NI_1000_YEARS_0_IBTRACSDELTA.txt')
tc_tracks_sync_extent = tracks_sync_1000yr.extent
```

```python
ax = tc_tracks.plot(adapt_fontsize=False,figsize=(13,9))
ax.set_extent(tc_tracks_sync_extent, crs=ccrs.PlateCarree())

plt.tight_layout()
plt.show()
```
