# STORM import and visualization

This notebook involves importing STORM synthetic database in a CLIMADA environment.

```python
from climada.hazard import TCTracks
import matplotlib.pyplot as plt
import matplotlib as mpl
import cartopy.crs as ccrs
```

```python
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "cmr10"
})
plt.rcParams['font.size'] = 12
mpl.rcParams['axes.unicode_minus'] = False
```

We import the first 1000 years of the STORM synthetic database.

```python
tracks_sync_1000yr = TCTracks.from_simulations_storm('../Data/STORM_DATA_HadGEM3-GC31-HM_NI_1000_YEARS_0_IBTRACSDELTA.txt')
```

```python
tc_tracks_sync_extent = tracks_sync_1000yr.extent
```

```python
ax = tracks_sync_1000yr.plot(adapt_fontsize=False, figsize=(13,9))
ax.set_extent(tc_tracks_sync_extent, crs=ccrs.PlateCarree())

plt.tight_layout()
plt.show()
```
