An example of a historical track and its synthetic tracks.

```python
from climada.hazard import TCTracks
import matplotlib as mpl
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "cmr10"
})
plt.rcParams['font.size'] = 22
mpl.rcParams['axes.unicode_minus'] = True
```

A TC event, officially named as 'Amphan', happened to NIO in 2020. It is used as an example to illustrate the relationship between a historical TC track and its synthetic tracks. The single historical track is plotted first.

```python
tr_amphan = TCTracks.from_ibtracs_netcdf(provider="usa", 
                                         storm_id="2020136N10088")

fig = plt.figure(figsize=(9, 13))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

tr_amphan.plot(axis=ax)

ax.set_extent([80, 100, 5, 30], crs=ccrs.PlateCarree())

gl = ax.gridlines(draw_labels=True, linestyle='-', linewidth=0, color='none')
gl.top_labels = False
gl.right_labels = False

plt.show()
```

Next, its synthetic tracks are plotted.

```python
tr_amphan = TCTracks.from_ibtracs_netcdf(provider="usa", 
                                         storm_id="2020136N10088")

tr_amphan.equal_timestep()
tr_amphan.calc_perturbed_trajectories(nb_synth_tracks=5)

fig = plt.figure(figsize=(9, 13))
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

tr_amphan.plot(axis=ax)

ax.set_extent([80, 100, 5, 30], crs=ccrs.PlateCarree())

gl = ax.gridlines(draw_labels=True, linestyle='-', linewidth=0, color='none')
gl.top_labels = False
gl.right_labels = False

plt.show()
```
