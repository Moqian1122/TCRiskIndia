```python
import pandas as pd
```

```python
financing_data = pd.read_excel('Global-Coal-Project-Finance-Tracker_October-2024_GEM.xlsx', sheet_name='Data', header = 1)
```

```python
financing_data.shape[0]
```

Global data is successfully loaded.

```python
financing_data.columns
```

We only want India data.

```python
financing_india = financing_data[financing_data['Country/Area']=='India']
```

```python
financing_india.shape[0]
```

We do not need so many columns. So we just selected appearantly useful ones.

```python
financing_india_select_columns = financing_india[['Financier', 'Financier Public or Private', 'Financier Type',
       'Domestic/International', 'Financier Country', 'Advisor',
       'Type of Finance', 'Equity Partnership', 'Units Included', 'Capacity (MW)',
       'Financing Status', 'GEM unit/phase ID', 'GEM location ID', 'Country/Area', 'Plant name', 'Status', 'Start year', 'Latitude',
       'Longitude']]
```

We consider operating/construction/permitted coal plants.

```python
financing_india_select_by_status = financing_india_select_columns[financing_india_select_columns['Status'].isin(['construction', 'operating', 'permitted'])]
```

```python
financing_india_select_by_status.shape[0]
```

In the end we want to group by asset. GEM location ID is unique so it could be the key.

```python
financing_india_asset = financing_india_select_by_status[financing_india_select_by_status.duplicated(subset=['GEM location ID'])==0]
```

```python
financing_india_asset
```

We want to have an Excel file exported from this. Before exporting, the dataset is reindexed strating from 0.

```python
financing_india_asset = financing_india_asset.reset_index(drop=True)
```

```python
financing_india_asset.to_excel('financing_asset_india.xlsx', sheet_name = 'Data')
```
